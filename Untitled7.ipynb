```python
df['Total_health']=0.5*((df['MentHlth'])+(df['PhysHlth']))
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['BMI'] = scaler.fit_transform(df[['BMI']])
df['Total_health'] = scaler.fit_transform(df[['Total_health']])
df['GenHlth'] = scaler.fit_transform(df[['GenHlth']])
```

```python
df_cleaned=df.drop(['MentHlth','PhysHlth','Income','CholCheck', 'Education','AnyHealthcare'], axis=1)
df_cleaned.head()
```

```python
num=df_cleaned.select_dtypes(include=['float64','int64'])
num.corr()
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(df_cleaned.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = df_cleaned.drop(columns=['Diabetes_binary'])
Y = df_cleaned['Diabetes_binary']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
# model training
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

```python
# print metric to get performance
from sklearn.neighbors import KNeighborsClassifier
model1 = KNeighborsClassifier()
```

```python
# knn - k-nearest neighbours
model1.fit(x_train, y_train)
print("Accuracy: ",model1.score(x_test, y_test) * 100)
```

```python
from sklearn.svm import SVC
model2=SVC()
```

```python
model2.fit(x_train, y_train)
print("Accuracy: ",model2.score(x_test, y_test) * 100)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(criterion='gini', max_depth=5, random_state=42)
```

```python
tree_model.fit(x_train, y_train)
print("Accuracy:", tree_model.score(x_test, y_test)*100)
```
